```python

import numpy as np
import pandas as pd
from pysurvey.plot import setup, line, legend, hist
```

```python
def _year(x):
    if not isinstance(x,str): return x
    year = 2000 + int(x[2:])
    if year > 2080: year -= 100
    qtr = ['WI','SP','SU', 'FA'].index(x[:2])*0.25
    return year+qtr

df = pd.read_csv('data/students.csv')
df['year'] = df['Admit Year'].apply(lambda x: _year(x.split('(')[1].split(',')[0]))
# df['left'] = (df['year'] + df['Years to PHD']).apply(lambda x: np.floor(x))
df['left'] = df['Withdrawal'].apply(lambda x: isinstance(x,str))
df['Years to Withdrawal'] = df['Withdrawal'].apply(lambda x: _year(x)) - df['year']
df[:3]
```

```python
x = df['year'].apply(int)
years = sorted(np.unique(x))
n = np.zeros(len(years))
m = np.zeros(len(years))
fracs = np.zeros(len(years))
errs = np.zeros(len(years))
for i,year in enumerate(years):
    ii = np.where((x == year) & (df['Gender'] == 'F'))[0]
    jj = np.where((x == year) )[0]
    n[i] = len(jj)
    m[i] = len(ii)
    fracs[i] = len(ii)*1.0/len(jj)
    tot = len(ii)+len(jj)
    errs[i] = np.sqrt(fracs[i]*(1-fracs[i])/tot)

setup(figsize=(12,6), subplt=(1,2,1), xlabel='year')
pylab.step(years, n)
pylab.step(years, m)
setup(subplt=(1,2,2), xlabel='year')
pylab.errorbar(years, fracs,errs, fmt='+')



    
```
